```python
#### Normalization ####
def min_max(data):
    min=data.min(axis=0)
    max=data.max(axis=0)
    norm=(data-min)/(max-min)
    return norm
```

```python
#### Low pass filter ####    
def LPF2(data):
    Data = np.zeros((data.shape[0],data.shape[1]))
    y = np.zeros(data.shape[0])
    for s in range(data.shape[1]): #channel
        for t in range(data.shape[0]):
            y[t] = data[t,s]
        N=len(y)
        yf = np.fft.fft(y)/(N/2)
        yf[lpf:N]=0
        y2 = np.fft.ifft(yf)*(N/2)
        for t in range(data.shape[0]):
            Data[t,s] = y2[t]
    return Data
```

```python
#### 学習用正常データバッチ作成 ####
def make_batch(data,batch_size,augumentation): 
    batch = np.zeros((batch_size, timesteps, num_input)) #num_input(62行目)
    batch = np.array(batch, dtype=np.float32)
    output = np.zeros((batch_size, num_input))
    output = np.array(output, dtype=np.float32)
    aug_batch = np.zeros((augumentation,  timesteps, num_input))
    aug_batch = np.array(aug_batch, dtype=np.float32)
    aug_output = np.zeros((augumentation, num_input))
    aug_output = np.array(aug_output, dtype=np.float32)
    for h in range(batch_size): #batch_size = 20 (51行目)
        aug_rand = random.randint(0,augumentation-1) #augumentation(54行目)
        index = random.randint(train_start+timesteps, train_size) #train_size(68行目）学習データセット数　＜1008
         
        for i in range(augumentation):
            #rand = random.uniform(0,2*fluctuation)#一様分布乱数 rand=0~0.2, fluctuation(53行目)
                       
            for k in range(num_input): # numinput=読み込みデータの次元
                #aug_output[i, k] = (1-fluctuation+rand)*data[index + timesteps][k]
                rand = random.uniform(0,2*fluctuation)#一様分布乱数 rand=0~0.2, fluctuation(53行目)
                aug_output[i, k] = (1-fluctuation+rand)*data[index][k]
                            #10stepの各学習データの次の点 #fluctuation+rand= -0.1~0.1
            for k in range(num_input):
                for j in range(timesteps): #timesteps(61行め)
                    #aug_batch[i, j, k] = (1-fluctuation+rand)*data[index + j][k]
                    rand = random.uniform(0,2*fluctuation)#一様分布乱数 rand=0~0.2, fluctuation(53行目)
                    aug_batch[i, j, k] = (1-fluctuation+rand)*data[index -timesteps + j][k]
        output[h] = aug_output[aug_rand]
        batch[h] = aug_batch[aug_rand]
    return batch, output
```

```python
#モデル
class MyModel(tf.keras.Model):
    def __init__(self):
        he_initializer = tf.keras.initializers.he_normal()
        norm_initializer = tf.keras.initializers.RandomNormal(mean=1.0, stddev=0.01)
        super(MyModel, self).__init__()
        self.lstm = layers.LSTM(units=128, return_sequence=True,input_shape=(40,200,12))
        self.lstm = layers.Dropout(0.2)
#kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu', input_shape=        
        self.norm1 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d1 = layers.Conv1D(128, kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu')
        self.norm2 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d2 = layers.Conv1D(1, kernel_size=(1,), kernel_initializer=he_initializer, activation='sigmoid')
        self.pooling = layers.AveragePooling1D(pool_size=10)

    def call(self, x):
#         x = tf.reshape(x, [-1,20,12])
        x = self.lstm(x)
        x = self.norm1(x)
        x = self.conv1d1(x)
        x = self.norm2(x)
        x = self.conv1d2(x)
        x = self.pooling(x)
        return x

model = MyModel()
model.build(input_shape=(40,200,12))
model.summary()
#optimizer = tf.keras.optimizers.RMSprop(0.001)　#最適なoptimizer、パラメータを探す必要あり
optimizer = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
```
